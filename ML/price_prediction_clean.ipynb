```python
import pandas as pd
import re
from fuzzywuzzy import process

# Expanded mapping of Arabic words to numbers (always store without "ال")
arabic_numbers = {
    "أرضي": 0, "ارض": 0, "ارضية": 0, "ارضي": 0, "أرضى": 0,
    "اول": 1, "أول": 1,
    "ثاني": 2, "ثانى": 2, "ثان": 2, "تاني": 2, "تانى": 2,
    "ثالث": 3, "تالت": 3, "ثالت": 3,
    "رابع": 4, 
    "خامس": 5, 
    "سادس": 6, 
    "سابع": 7, 
    "ثامن": 8, "تامن": 8,
    "تاسع": 9, 
    "عاشر": 10, 
    "حادي عشر": 11, "حادى عشر": 11, "حداشر": 11,
    "ثاني عشر": 12, "ثانى عشر": 12, "اتناشر": 12,
    "ثالث عشر": 13, "تالت عشر": 13, "تلاتاشر": 13,
    "رابع عشر": 14, "اربعتاشر": 14,
    # "روف": "roof"
}

# Convert Arabic numbers to Western (١ → 1, ٢ → 2)
arabic_to_western_digits = str.maketrans("٠١٢٣٤٥٦٧٨٩", "0123456789")

# Create two versions of the mapping - one with "ال" and one without
normalized_keys = {re.sub(r'^ال', '', key): value for key, value in arabic_numbers.items()}
normalized_keys_with_al = {f"ال{key}": value for key, value in arabic_numbers.items() if not key.startswith("ال")}
# Combine both dictionaries
all_keys = {**normalized_keys, **normalized_keys_with_al}

def normalize_arabic(text):
    """
    Normalize Arabic text: remove diacritics, convert similar letters.
    Note: We don't remove "ال" here as we need to keep it for specific patterns.
    """
    if not isinstance(text, str):
        return ""
    
    text = re.sub(r'[\u0617-\u061A\u064B-\u0652]', '', text)  # Remove diacritics
    text = re.sub(r'[إأآ]', 'ا', text)  # Normalize Alef
    text = re.sub(r'ة', 'ه', text)  # Normalize Teh Marbuta
    text = re.sub(r'ى', 'ي', text)  # Normalize Alef Maqsura to Yaa
    text = text.translate(arabic_to_western_digits)  # Convert Arabic numbers to Western
    return text.strip()

def get_best_match(word, include_al=False):
    """
    Finds the best match for a possibly misspelled floor name using fuzzy matching.
    """
    keys_to_check = all_keys if include_al else normalized_keys
    
    best_match, score = process.extractOne(word, keys_to_check.keys())
    return keys_to_check[best_match] if score > 80 else None

def extract_floors(description):
    """
    Extracts floor numbers from a description string.
    Returns a list of detected floors.
    """
    if not isinstance(description, str):
        return []

    # Normalize the description but preserve "ال"
    normalized_desc = normalize_arabic(description)
    
    # List to store extracted floors
    extracted_floors = []
    
    # Handle the specific format with colon "الدور: الثالث" or "الطابق: 5"
    colon_patterns = [
        r'الدور\s*[:]\s*(ال?\w+|\d+)',
        r'طابق\s*[:]\s*(ال?\w+|\d+)',
        r'الطابق\s*[:]\s*(ال?\w+|\d+)'
    ]
    
    for pattern in colon_patterns:
        colon_matches = re.findall(pattern, normalized_desc, re.IGNORECASE)
        
        for match in colon_matches:
            # If it's a digit, convert directly
            if match.isdigit():
                extracted_floors.append(int(match))
                continue
                
            # Check if match is in our dictionaries (with or without "ال")
            if match in all_keys:
                extracted_floors.append(all_keys[match])
            else:
                # Try fuzzy matching with "ال" preserved
                best_match = get_best_match(match, include_al=True)
                if best_match is not None:
                    extracted_floors.append(best_match)
    
    # Patterns to match different floor expressions
    patterns = [
        # Standard forms with "دور" or "طابق" followed by a number or word
        r'(?:بالدور|الدور|دور|طابق|الطابق)\s*(?:ال)?\s*([\d\u0660-\u0669\u0600-\u06FF]+)',
        
        # Standalone floor descriptions
        r'\s(ارضي|أرضي|ارضى|أرضى|اول|أول|ثاني|ثانى|ثان|تاني|تانى|ثالث|تالت|رابع|خامس|سادس|سابع|ثامن|تامن|تاسع|عاشر|حادي عشر|ثاني عشر|روف)\s',
        
        # Floor as part of property description
        r'شقه\s+(ارضي|أرضي|ارضى|أرضى|اول|أول|ثاني|ثانى|ثان|تاني|تانى|ثالث|تالت|رابع|خامس|سادس|سابع|ثامن|تامن|تاسع|عاشر|حادي عشر|ثاني عشر|روف)',
        r'شقة\s+(ارضي|أرضي|ارضى|أرضى|اول|أول|ثاني|ثانى|ثان|تاني|تانى|ثالث|تالت|رابع|خامس|سادس|سابع|ثامن|تامن|تاسع|عاشر|حادي عشر|ثاني عشر|روف)',
        
        # Floor in a different format
        r'بالدور\s+ال\s*(\d+)',
        
        # Matching "دور ثانى"
        r'\s+دور\s+(ارضي|أرضي|ارضى|أرضى|اول|أول|ثاني|ثانى|ثان|تاني|تانى|ثالث|تالت|رابع|خامس|سادس|سابع|ثامن|تامن|تاسع|عاشر|حادي عشر|ثاني عشر|روف)'
    ]
    
    # Apply each pattern to find matches
    for pattern in patterns:
        matches = re.findall(pattern, ' ' + normalized_desc + ' ', re.IGNORECASE)
        for match in matches:
            # If it's a digit, convert directly
            if isinstance(match, str) and match.isdigit():
                extracted_floors.append(int(match))
                continue
                
            # Try to find exact match in our normalized keys
            match_norm = match.strip()
            if match_norm in normalized_keys:
                extracted_floors.append(normalized_keys[match_norm])
            else:
                # Use fuzzy matching for typos and variations
                best_match = get_best_match(match_norm)
                if best_match is not None:
                    extracted_floors.append(best_match)
    
    # Remove duplicates while preserving order
    # unique_floors = []
    # for floor in extracted_floors:
    #     if floor not in unique_floors:
    #         unique_floors.append(floor)
    floors = extracted_floors[0] if len(extracted_floors) > 0 else None
            
    return floors

def process_dataframe(df):
    """
    Process the dataframe to extract floor information.
    """
    # Copy DataFrame to avoid modifying original
    df_processed = df.copy()
    
    # Apply extract_floors function to create a new "FLOORS" column
    df_processed["FLOORS"] = df_processed["DESCRIPTION"].apply(extract_floors)
    
    # Create a filter for apartments (شقة) that mention floors
    has_floor_mention = df_processed["DESCRIPTION"].str.contains(
        "طابق|دور|الدور|الطابق|ارضي|أرضي|ارضى|أرضى", 
        na=False, 
        regex=True
    )
    is_apartment = df_processed["PROPERTY_TYPE"] == "شقة"
    
    df_processed.loc[~is_apartment, "FLOORS"] = -1
    
    return df_processed

# Example usage:
df_cleaned = process_dataframe(df_cleaned)
df_cleaned
```

```python
import pandas as pd
import re
from fuzzywuzzy import process

# Mapping of Arabic number words for years (different from floor numbering)
arabic_year_numbers = {
    "سنة": 1, "سنه": 1, "عام": 1, "سنوية": 1,
    "سنتين": 2, "عامين": 2,
    "ثلاث": 3, "تلات": 3, "ثلاثة": 3, "تلاته": 3, "تلاتة": 3,
    "اربع": 4, "أربع": 4, "اربعة": 4, "أربعة": 4, "اربعه": 4, "أربعه": 4,
    "خمس": 5, "خمسة": 5, "خمسه": 5,
    "ست": 6, "ستة": 6, "سته": 6, 
    "سبع": 7, "سبعة": 7, "سبعه": 7,
    "ثمان": 8, "ثمانية": 8, "ثمانيه": 8, "تمان": 8, "تمانية": 8, "تمانيه": 8,
    "تسع": 9, "تسعة": 9, "تسعه": 9,
    "عشر": 10, "عشرة": 10, "عشره": 10
}

# Convert Arabic numbers to Western (١ → 1, ٢ → 2)
arabic_to_western_digits = str.maketrans("٠١٢٣٤٥٦٧٨٩", "0123456789")

# Create normalized keys for better matching
normalized_year_keys = {re.sub(r'^ال', '', key): value for key, value in arabic_year_numbers.items()}

def normalize_arabic(text):
    """
    Normalize Arabic text: remove diacritics, convert similar letters.
    """
    if not isinstance(text, str):
        return ""
    
    text = re.sub(r'[\u0617-\u061A\u064B-\u0652]', '', text)  # Remove diacritics
    text = re.sub(r'[إأآ]', 'ا', text)  # Normalize Alef
    text = re.sub(r'ة', 'ه', text)  # Normalize Teh Marbuta
    text = re.sub(r'ى', 'ي', text)  # Normalize Alef Maqsura to Yaa
    text = text.translate(arabic_to_western_digits)  # Convert Arabic numbers to Western
    return text.strip()

def get_best_year_match(word):
    """
    Finds the best match for a possibly misspelled year number using fuzzy matching.
    """
    best_match, score = process.extractOne(word, normalized_year_keys.keys())
    return normalized_year_keys[best_match] if score > 80 else None

def extract_installment_period(description):
    """
    Extracts installment period (in years) from a description string.
    Returns the number of years for installment or None if not found.
    """
    if not isinstance(description, str):
        return None

    # Normalize the description
    normalized_desc = normalize_arabic(description)
    
    # Add spaces around description for better pattern matching
    padded_desc = ' ' + normalized_desc + ' '
    
    # Check if installment is mentioned - including the new cases
    installment_terms = [
        "قسط", "اقساط", "تقسيط", "بالتقسيط", "قسط علي", "قسط على", 
        "اقساط علي", "اقساط على", "بالاقساط", "بالأقساط",
        "والباقي علي", "والباقي على", "والباقى علي", "والباقى على",
        "والباقي", "والباقى", "الباقي علي", "الباقي على", "الباقى علي", "الباقى على"
    ]
    
    has_installment = any(term in padded_desc for term in installment_terms)
    if not has_installment:
        return 0
    
    # Patterns to match installment periods
    patterns = [
        # Direct number of years + year terms
        r'(\d+)\s*(سنوات|سنين|سنه|سنة|عام|اعوام|أعوام)',
        
        # Word number + year terms
        r'(ثلاث|تلات|اربع|أربع|خمس|ست|سبع|ثمان|تمان|تسع|عشر)\s*(سنوات|سنين|سنه|سنة|عام|اعوام|أعوام)',
        
        # Special case for "سنتين" (two years)
        r'(سنتين|عامين)',
        
        # Special case for "سنة" or "عام" (one year)
        r'(?:لمدة|مدة|على|علي|خلال)\s+(سنة|سنه|عام)(?:\s|$)',
        
        # Pattern for installment over years
        r'(?:قسط|اقساط|تقسيط|بالتقسيط)\s+(?:علي|على)\s+(\d+)\s*(?:سنوات|سنين|سنه|سنة|عام|اعوام|أعوام)',
        
        # Pattern for installment with word numbers
        r'(?:قسط|اقساط|تقسيط|بالتقسيط)\s+(?:علي|على)\s+(ثلاث|تلات|اربع|أربع|خمس|ست|سبع|ثمان|تمان|تسع|عشر)\s*(?:سنوات|سنين|سنه|سنة|عام|اعوام|أعوام)',
        
        # New patterns for "والباقي على" cases
        r'(?:والباقي|والباقى|الباقي|الباقى)\s+(?:علي|على)\s+(\d+)\s*(?:سنوات|سنين|سنه|سنة|عام|اعوام|أعوام)',
        
        # New patterns for "والباقي على" with word numbers
        r'(?:والباقي|والباقى|الباقي|الباقى)\s+(?:علي|على)\s+(ثلاث|تلات|اربع|أربع|خمس|ست|سبع|ثمان|تمان|تسع|عشر)\s*(?:سنوات|سنين|سنه|سنة|عام|اعوام|أعوام)',
        
        # Pattern for "والباقي" followed by year terms but without "على" explicitly
        r'(?:والباقي|والباقى|الباقي|الباقى)\s+(\d+)\s*(?:سنوات|سنين|سنه|سنة|عام|اعوام|أعوام)',
        
        # Pattern for "والباقي" with word numbers but without "على" explicitly
        r'(?:والباقي|والباقى|الباقي|الباقى)\s+(ثلاث|تلات|اربع|أربع|خمس|ست|سبع|ثمان|تمان|تسع|عشر)\s*(?:سنوات|سنين|سنه|سنة|عام|اعوام|أعوام)'
    ]
    
    for pattern in patterns:
        matches = re.findall(pattern, padded_desc, re.IGNORECASE)
        if matches:
            for match in matches:
                # Handle tuple results from regex groups
                if isinstance(match, tuple):
                    match = match[0]  # Get the first capturing group
                
                # If it's a digit, convert directly
                if match.isdigit():
                    return int(match)
                
                # Check direct matches in our year numbers dictionary
                if match in normalized_year_keys:
                    return normalized_year_keys[match]
                
                # Special cases
                if match in ["سنتين", "عامين"]:
                    return 2
                if match in ["سنة", "سنه", "عام"]:
                    return 1
                
                # Use fuzzy matching for variations
                best_match = get_best_year_match(match)
                if best_match is not None:
                    return best_match
    
    # If we found installment terms but couldn't extract a specific period
    # we'll return a default value of 0 to indicate "installment with unknown period"
    return 0

def process_dataframe_for_installments(df):
    """
    Process the dataframe to extract installment period information.
    """
    # Copy DataFrame to avoid modifying original
    df_processed = df.copy()
    
    # Apply extract_installment_period function to create a new "INSTALLMENT_YEARS" column
    df_processed["INSTALLMENT_YEARS"] = df_processed["DESCRIPTION"].apply(extract_installment_period)
    
    # Create a flag for properties with installment options
    df_processed["HAS_INSTALLMENT"] = df_processed["INSTALLMENT_YEARS"] > 0
    
    return df_processed

# Example usage:
df_cleaned = process_dataframe_for_installments(df_cleaned)
df_cleaned
```

```python
import pandas as pd
from scipy.stats import pointbiserialr, f_oneway

# Ensure PRICE is numeric
df_cleaned['PRICE'] = pd.to_numeric(df_cleaned['PRICE'], errors='coerce')
df_cleaned = df_cleaned.dropna(subset=['PRICE'])

# Store results
results = []

# Identify binary categorical columns correctly
binary_columns = [col for col in df_cleaned.columns if df_cleaned[col].nunique() == 2]

# Convert binary categorical variables to numeric (0 and 1)
for col in binary_columns:
    df_cleaned[col] = df_cleaned[col].astype('category').cat.codes
    corr, p_value = pointbiserialr(df_cleaned[col], df_cleaned['PRICE'])
    results.append((col, 'Point Biserial', corr, p_value))

# Identify multi-level categorical columns
multi_level_columns = [col for col in df_cleaned.select_dtypes(include=['object']).columns if df_cleaned[col].nunique() > 2]

# ANOVA for multi-level categorical variables
for col in multi_level_columns:
    try:
        groups = [df_cleaned.loc[df_cleaned[col] == level, 'PRICE'] for level in df_cleaned[col].unique()]
        if all(len(group) > 1 for group in groups):  # Ensure each group has >1 sample
            anova_result = f_oneway(*groups)
            results.append((col, 'ANOVA', anova_result.statistic, anova_result.pvalue))
    except Exception as e:
        print(f"Skipping {col} due to error: {e}")

# Sort results by absolute correlation/statistic value
sorted_results = sorted(results, key=lambda x: abs(x[2]), reverse=True)

# Print sorted results
for col, test_type, value, p_value in sorted_results:
    print(f'{col} ({test_type}): Value: {value:.4f}, p-value: {p_value:.4e}')
```
